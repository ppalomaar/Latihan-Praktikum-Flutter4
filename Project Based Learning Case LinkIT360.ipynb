# 2. Handling Missing Columns

```python
#  Hitung median untuk renewal_date
median_date = df['renewal_date'].median()
#  Mengisi missing values pada renewal_date dengan median
df['renewal_date'] = df['renewal_date'].fillna(median_date)
#  Mengisi missing values pada adnet (tipe string)
df['adnet'].fillna('NA', inplace=True)
# Mengisi missing values pada unsubs_date dengan renewal_date
df['unsubs_date'] = df['unsubs_date'].fillna(df['renewal_date'])
df['publisher'] = df['publisher'].fillna('Nan')


missing_values_after = df.isnull().sum()
print(missing_values_after)
```

## 2.Adding new columns

```python
# Membuat kolom 'campaign' berdasarkan kondisi adnet dan status
df['campaign'] = df.apply(lambda row:
                          f"{row['adnet']}_Renewed" if row['status'] == 1 else
                          f"{row['adnet']}_Not_Renewed" if row['status'] == 0 else
                          f"{row['adnet']}_Unsubscribed", axis=1)

# membuat kolom 'subscription_order_months' untuk mengetahui rentang berlangganan
df['subscription_tenure_months'] = ((df['unsubs_date'] - df['subs_date']).dt.days) // 30

# Menghitung AOV
df['average_order_value'] = np.where(df['success_billing'] > 0,
                                     df['revenue'] / df['success_billing'],
                                     np.nan)  # Menghasilkan NaN jika success_billing = 0

# Mengisi nilai NaN dengan 0 (opsional, tergantung kebutuhan)
df['average_order_value'].fillna(0, inplace=True)

# membuat kolom sales per customer
total_revenue = df['revenue'].sum()
distinct_trx_count = df['trxid'].nunique()

df['sales_per_customer'] = total_revenue / distinct_trx_count

# Menampilkan DataFrame
print(df)
```

## 2.handling outliers

```python
boxplot_numeric_columns = ['revenue', 'service_price', 'attempt_charging', 'success_billing', 'average_order_value', 'sales_per_customer', 'subscription_tenure_months']

# Membuat boxplot untuk setiap kolom
plt.figure(figsize=(12, 8))

for i, column in enumerate(boxplot_numeric_columns):
    plt.subplot(3, 3, i+1)  # Membuat subplot 3x3
    sns.boxplot(y=df[column])
    plt.title(f'Boxplot {column}')

plt.tight_layout()
plt.show()
```

```python
# Kolom numerik yang akan diperiksa
normalize_numeric_columns = ['revenue', 'service_price', 'success_billing', 'subscription_tenure_months']

outliers_summary = {}

# Fungsi winsorize_series
def winsorize_series(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return series.clip(lower=lower_bound, upper=upper_bound)

# Loop untuk setiap kolom numerik
for column in normalize_numeric_columns:
    # Menghitung Q1, Q3, IQR, dan batas outliers
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menghitung jumlah outliers
    data_without_outliers = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    outliers_count = df.shape[0] - data_without_outliers.shape[0]

    # Menambahkan informasi ke dictionary outliers_summary
    outliers_summary[column] = {
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'Lower Bound': lower_bound,
        'Upper Bound': upper_bound,
        'Outliers Count': outliers_count,
        'Outliers Percentage': (outliers_count / df.shape[0]) * 100
    }

# Menampilkan ringkasan outliers
print(outliers_summary)

# Menerapkan winsorizing pada kolom yang memiliki outliers
# df['attempt_charging'] = winsorize_series(df['attempt_charging'])
# df['success_billing'] = winsorize_series(df['success_billing'])

# Mengecek summary setelah winsorizing
winsorized_summary = df[['revenue', 'service_price', 'success_billing', 'subscription_tenure_months']].describe()
print(winsorized_summary)
```

# EDA

## 1.Churn risk score based on engagement

```python
# menduplikasi dataframe untuk mengganti tipe data
df_churn = df_normalized.copy()
df_churn['status'] = pd.to_numeric(df['status'], errors='coerce')
df_churn.head()
```

```python
# Menghitung median dari setiap kolom
median_success_billing = df_churn['success_billing'].median()
median_charging_attempts = df_churn['attempt_charging'].median()
median_status = df_churn['status'].median()

# Menentukan risiko churn berdasarkan kondisi median
def classify_risk(row):
  if (
    (row['success_billing'] <= median_success_billing) &
    (row['attempt_charging'] <= median_charging_attempts) &
    (row['status'] <= median_status)
):
   return 'under_churn_risk'
  else:
   return 'over_churn_risk'

df_churn['churn_risk_category'] = df.apply(classify_risk, axis=1)

# Menampilkan DataFrame dengan kolom baru
print(df_churn.head())
```

```python
# Sample data for demonstration (replace with your DataFrame values)
categories = ['Success Billing', 'Charging Attempts', 'Status']
under_churn_risk = [
    df_churn[df_churn['churn_risk_category'] == 'under_churn_risk']['success_billing'].count(),
    df_churn[df_churn['churn_risk_category'] == 'under_churn_risk']['attempt_charging'].count(),
    df_churn[df_churn['churn_risk_category'] == 'under_churn_risk']['status'].count()
]

upper_churn_risk = [
    df_churn[df_churn['churn_risk_category'] == 'over_churn_risk']['success_billing'].count(),
    df_churn[df_churn['churn_risk_category'] == 'over_churn_risk']['attempt_charging'].count(),
    df_churn[df_churn['churn_risk_category'] == 'over_churn_risk']['status'].count()
]

# Define bar width and positions
bar_width = 0.50
x = np.arange(len(categories))

# Create the plot
plt.figure(figsize=(10, 6))
plt.bar(x - bar_width/2, under_churn_risk, width=bar_width, label='Under Churn Risk', color='skyblue')
plt.bar(x + bar_width/2, upper_churn_risk, width=bar_width, label='Upper Churn Risk', color='navy')

# Add labels, title, and legend
plt.xlabel('Engagement Metrics')
plt.ylabel('Count')
plt.title('Churn Risk Comparison')
plt.xticks(x, categories)  # Set x-axis labels
plt.legend()

# Display the plot
plt.show()
```

```python
df_churn['subs_date'] = pd.to_datetime(df['subs_date'])

# Menentukan periode awal (bulan pertama berlangganan)
cohort_start_date = df_churn['subs_date'].min()
cohort_end_date = cohort_start_date + pd.DateOffset(months=1)

# Mengambil pengguna yang berlangganan di bulan pertama
first_month_users = df_churn[(df_churn['subs_date'] >= cohort_start_date) & (df_churn['subs_date'] < cohort_end_date)]

# Menentukan pengguna yang masih aktif setelah 1 bulan
retained_users = first_month_users[first_month_users['status'] == 1]

# Menghitung tingkat retensi
retention_rate = len(retained_users) / len(first_month_users) * 100

# Menampilkan hasil
df_churn['retention_rate'] = retention_rate
print(f"Tingkat retensi setelah 1 bulan: {retention_rate:.2f}%")
```

```python
# Visualisasi scatter plot dengan failed_charge_attempts_ratio, retention_rate, dan revenue
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_churn,
    x='failed_charging_attempts_ratio',  # Failed charge ratio
    y='retention_rate',  # Retention rate
    sizes=(20, 200),  # Ukuran titik dapat bervariasi
    palette='viridis',  # Skema warna untuk revenue
    alpha=0.7
)

# Menambahkan label dan judul
plt.title('Scatterplot of user retention on Failed Charge Ratio', fontsize=16)
plt.xlabel('Failed Charge Attempts Ratio', fontsize=12)
plt.ylabel('Retention Rate (%)', fontsize=12)
plt.show()
```
